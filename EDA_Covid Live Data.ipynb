### Corona Virus Live Tracking For India

```python
import pandas as pd
df = pd.read_json("https://www.mohfw.gov.in/data/datanew.json")
df.head()
```

```python
df
```

```python
df.columns
```

```python
df = df.iloc[:-1,1:]
```

```python
df = df.drop(['death_reconsille','total'], axis = 1)
```

```python
df.tail()
```

```python
import numpy as np
total_states = np.arange(len(df['state_name']))
```

```python
total_states
```

```python
df['active']
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

### Total Positive Cases Based On States In India

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(10, 8), dpi=80, facecolor='w', edgecolor='k')
plt.barh(total_states, df['positive'], align= 'center', color = (1,0,0), alpha = 0.5, edgecolor=(0.5,0.2,0.8))
plt.yticks(total_states, df['state_name'])  
plt.xlim(1,max(df['positive'])+100000) 
plt.xlabel('Positive Number of Cases')  
plt.title('Corona Virus Cases')  
plt.show()  
```

### Active Number of Cases based on States

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(10, 8), dpi=80, facecolor='w', edgecolor='k')
plt.barh(total_states, df['active'], align= 'center', color = (0,0,1), alpha = 0.5, edgecolor=(0.5,0.2,0.8))
plt.yticks(total_states, df['state_name'])  
plt.xlim(1,max(df['active'])+10) 
plt.xlabel('Active Number of Cases')  
plt.title('Corona Virus Cases')  
plt.show() 
```

### Number of Deaths based on States

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(10, 8), dpi=80, facecolor='w', edgecolor='k')
plt.barh(total_states, df['death'], align= 'center', color = (1,0.5,1), alpha = 0.5, edgecolor=(0.5,0.2,0.8))
plt.yticks(total_states, df['state_name'])  
plt.xlim(1,max(df['death'])+1000) 
plt.xlabel('No of Deaths')  
plt.title('Corona Virus Cases')  
plt.show()
```
